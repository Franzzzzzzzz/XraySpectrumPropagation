```python
import spekpy as sp
import numpy as np
import matplotlib.pyplot as plt

Dtd = 100 # distance tube detector in cm
s = sp.Spek(kvp=180,th=20, dk=0.5, z=Dtd, targ='W', char=False) # Generate a spectrum. char: include characteristics freq or not
s.filter('Be', 0.8) # Filter by 0.8 mm of Be

[freq, fluence] = s.get_spectrum()
plt.plot(freq,fluence, '.')

print(len(freq))
```
